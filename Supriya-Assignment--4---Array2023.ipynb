```python
#<aside>
ðŸ’¡ **Question 1**
Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** 
the integers that appeared in **all** three arrays.

**Example 1:**

Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

Output: [1,5]

**Explanation:** Only 1 and 5 appeared in the three arrays.

</aside>

#To find the integers that appeared in all three arrays, you can use a common element search algorithm. Here's an 
implementation in Python:

def find_common_elements(arr1, arr2, arr3):
    result = []
    i = j = k = 0

    while i < len(arr1) and j < len(arr2) and k < len(arr3):
        if arr1[i] == arr2[j] == arr3[k]:
            result.append(arr1[i])
            i += 1
            j += 1
            k += 1
        elif arr1[i] < arr2[j]:
            i += 1
        elif arr2[j] < arr3[k]:
            j += 1
        else:
            k += 1

    return result

#Using the given example:

arr1 = [1,2,3,4,5]
arr2 = [1,2,5,7,9]
arr3 = [1,3,4,5,8]

common_elements = find_common_elements(arr1, arr2, arr3)
print(common_elements)  # Output: [1, 5]

"""The algorithm iterates through the three arrays simultaneously and compares the elements at the current indices. If the 
elements are the same, they are added to the result list. If the element in arr1 is smaller than the element in arr2, the 
index of arr1 is incremented. If the element in arr2 is smaller than the element in arr3, the index of arr2 is incremented. 
Otherwise, the index of arr3 is incremented. This process continues until the end of any of the arrays is reached. 
The resulting result list contains the integers that appeared in all three arrays."""
```

```python
#<aside>
ðŸ’¡ **Question 2**

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2.
Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. 
Therefore, answer[1] = [4,6].

</aside>

#To find the distinct integers in nums1 that are not present in nums2 and vice versa, you can use set operations.
Here's an implementation in Python:

def find_missing_elements(nums1, nums2):
    set_nums1 = set(nums1)
    set_nums2 = set(nums2)

    missing_nums1 = list(set_nums1 - set_nums2)
    missing_nums2 = list(set_nums2 - set_nums1)

    return [missing_nums1, missing_nums2]

#Using the given example:

nums1 = [1, 2, 3]
nums2 = [2, 4, 6]

missing_elements = find_missing_elements(nums1, nums2)
print(missing_elements)  # Output: [[1, 3], [4, 6]]

"""The algorithm converts nums1 and nums2 into sets using the set() function to remove duplicates. Then, set operations are 
performed to find the elements that are present in one set but not in the other. The set_nums1 - set_nums2 operation finds
the elements in set_nums1 that are not in set_nums2, and the set_nums2 - set_nums1 operation finds the elements in set_nums2 
that are not in set_nums1. Finally, the resulting sets are converted back to lists and returned as the answer."""
```

```python
#<aside>
ðŸ’¡ **Question 3**
Given a 2D integer array matrix, return *the **transpose** of* matrix.

The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

**Example 1:**

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [[1,4,7],[2,5,8],[3,6,9]]

</aside>

#To find the transpose of a matrix, you need to switch the row and column indices of each element. Here's a Python function 
that can accomplish this:

def transpose(matrix):
    # Get the dimensions of the input matrix
    rows = len(matrix)
    cols = len(matrix[0])

    # Create a new matrix with swapped dimensions
    transposed = [[0 for _ in range(rows)] for _ in range(cols)]

    # Populate the transposed matrix
    for i in range(rows):
        for j in range(cols):
            transposed[j][i] = matrix[i][j]

    return transposed

#Using the example you provided:


matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed_matrix = transpose(matrix)
print(transposed_matrix)

"""The function takes the input matrix and creates a new matrix with swapped dimensions. It then iterates through each
element in the input matrix and assigns it to the corresponding position in the transposed matrix. Finally, it returns
the transposed matrix."""
```

```python
#<aside>
ðŸ’¡ **Question 4**
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that 
the sum of min(ai, bi) for all i is **maximized**. Return *the maximized sum*.

**Example 1:**

Input: nums = [1,4,3,2]

Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3

2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3

3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

So the maximum possible sum is 4.

</aside>

#To maximize the sum of the minimum elements in each pair, we should pair the smallest numbers with each other. Sorting the
array in ascending order will help us achieve this. Once the array is sorted, we can pair the adjacent elements to get the 
maximum sum.

#Here's a Python function that implements this approach:

def array_pair_sum(nums):
    nums.sort()  # Sort the array in ascending order
    max_sum = 0

    for i in range(0, len(nums), 2):
        max_sum += nums[i]

    return max_sum

#Using the example you provided:

nums = [1, 4, 3, 2]
max_sum = array_pair_sum(nums)
print(max_sum)

"""The function sorts the input array nums in ascending order. Then, it iterates over the array from index 0 to the
second-to-last index, incrementing the index by 2 in each iteration. In each iteration, it adds the value at the current
index to the max_sum variable. Finally, it returns the max_sum as the maximum possible sum of the minimum elements in each 
pair."""
```

```python
#<aside>
ðŸ’¡ **Question 5**
You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has
exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**

[]()

![v2.jpg](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/4bd91cfa-d2b1-47b3-8197-a72e8dcfff4b/v2.jpg)

**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.

</aside>

"""To determine the number of complete rows in the staircase, we need to find the maximum value of k such that the sum of the 
first k natural numbers is less than or equal to n.

The sum of the first k natural numbers can be calculated using the formula: sum = (k * (k + 1)) / 2.

We can use this formula to find the maximum value of k by iterating from 1 to n and checking if the sum of the first k natural 
numbers exceeds n. Once we find a value of k for which the sum exceeds n, we return k - 1 as the number of complete rows."""

#Here's the implementation in Python:

def count_complete_rows(n):
    k = 1
    while (k * (k + 1)) // 2 <= n:
        k += 1
    return k - 1

#Let's test the example you provided:

n = 5
print(count_complete_rows(n))  # Output: 2

#The output is 2, indicating that there are 2 complete rows in the staircase for n = 5.
```

```python
#<aside>
ðŸ’¡ **Question 6**
Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** 
sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]

</aside>

"""To solve this problem, we can use a two-pointer approach. We can initialize two pointers, one pointing to the start of the
array (left pointer) and the other pointing to the end of the array (right pointer). We compare the absolute values of the
elements at the left and right pointers, square the larger absolute value, and place it at the end of a new result array.
We then move the corresponding pointer inward and repeat the process until the pointers meet."""

#Here's the implementation in Python:

def sorted_squares(nums):
    n = len(nums)
    result = [0] * n
    left = 0
    right = n - 1
    index = n - 1

    while left <= right:
        if abs(nums[left]) > abs(nums[right]):
            result[index] = nums[left] ** 2
            left += 1
        else:
            result[index] = nums[right] ** 2
            right -= 1
        index -= 1

    return result

#Let's test the example you provided:

nums = [-4, -1, 0, 3, 10]
print(sorted_squares(nums))  # Output: [0, 1, 9, 16, 100]

"""The output is [0, 1, 9, 16, 100], which is the array of the squares of each number sorted in non-decreasing order as
expected."""
```

```python
#<aside>
ðŸ’¡ **Question 7**
You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y]
should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

Count and return *the number of maximum integers in the matrix after performing all the operations*

**Example 1:**

![q4.jpg](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/4d0890d0-7bc7-4f59-be8e-352d9f3c1c52/q4.jpg)

**Input:** m = 3, n = 3, ops = [[2,2],[3,3]]

**Output:** 4

**Explanation:** The maximum integer in M is 2, and there are four of it in M. So return 4.

</aside>

"""To solve this problem, we can simulate the operations on the matrix and count the maximum integers.

Here's the step-by-step approach:

   1. Initialize the matrix M of size m x n with all zeros.

   2. Iterate through the ops array and for each operation op, update the matrix M by incrementing the values in the range
      (0, 0) to (op[0], op[1]) by one.

   3. After performing all the operations, iterate through the matrix M and find the maximum value.

   4. Count the number of occurrences of the maximum value in the matrix M.

   5. Return the count as the result."""

#Here's the Python code implementation for the above approach:

def maxCount(m: int, n: int, ops: List[List[int]]) -> int:
    # Step 1: Initialize the matrix M with all zeros
    M = [[0] * n for _ in range(m)]

    # Step 2: Perform the operations
    for op in ops:
        for i in range(op[0]):
            for j in range(op[1]):
                M[i][j] += 1

    # Step 3: Find the maximum value in M
    max_val = 0
    for i in range(m):
        for j in range(n):
            max_val = max(max_val, M[i][j])

    # Step 4: Count the occurrences of the maximum value
    count = 0
    for i in range(m):
        for j in range(n):
            if M[i][j] == max_val:
                count += 1

    # Step 5: Return the count
    return count

print(maxCount(3, 3, [[2,2],[3,3]]))

#In the given example, calling maxCount(3, 3, [[2,2],[3,3]]) will return 4, which is the expected output.
```

```python
#<aside>
ðŸ’¡ **Question 8**

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

*Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

**Example 1:**

**Input:** nums = [2,5,1,3,4,7], n = 3

**Output:** [2,3,5,4,1,7]

**Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

</aside>

"""To solve this problem, we can use a simple approach by creating a new array and interleaving the elements from the given
array nums based on the given pattern [x1, y1, x2, y2, ..., xn, yn].

Here's the step-by-step approach:

   1. Create an empty list result to store the interleaved elements.

   2. Iterate i from 0 to n (exclusive) using a for loop. This will ensure that we consider the elements at indices
      [0, 1, 2, ..., n-1].

   3. Inside the loop, append nums[i] to result to add the corresponding xi element.

   4. Append nums[i+n] to result to add the corresponding yi element.

   5. After the loop, result will contain the interleaved array in the required form.

   6. Return result as the final output."""

#Here's the Python code implementation for the above approach:

def shuffle(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])
        result.append(nums[i+n])
    return result

print(shuffle([2,5,1,3,4,7], 3))

#In the given example, calling shuffle([2,5,1,3,4,7], 3) will return [2, 3, 5, 4, 1, 7], which is the expected output.
```
